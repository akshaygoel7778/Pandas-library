```python
def red_cell(cell):
    if cell=='n.a.':
        return 'sam walton'
    else:
        return cell

df=pd.read_excel('stock_data.xlsx','Sheet1',converters={'people':red_cell})
```

```python
df
```

```python
df.to_excel('newfileexcel.xlsx',sheet_name='stocks',startrow=3,startcol=3,index=False)
```

```python
with pd.ExcelWriter('newfileexcel.xlsx') as writer:
    df.to_excel(writer,sheet_name="hg")
    df.to_excel(writer,sheet_name="yhy")
```
