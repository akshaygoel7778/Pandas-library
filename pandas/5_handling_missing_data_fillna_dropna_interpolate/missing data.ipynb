```python
import pandas as pd
```

```python
df=pd.read_csv('weather_data.csv',parse_dates=['day'])
```

```python
df
```

```python
df.set_index('day',inplace=True)
```

```python
df
```

```python
df.fillna(0)
```

```python
df.fillna({'temperature':0,'windspeed':0,'event':'no event'})
```

```python
df.fillna(method='ffill')# carry forwards the oprevious value
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(method='pad')
```

```python
df.fillna(method='bfill',axis='columns')
```

```python
df.fillna(method='ffill',limit=1) ## only firstnull value is filled
```

```python
df.interpolate() ## linear interpolation
```

```python
df.interpolate(method='time') ## linear interpolation
```

```python
df.dropna(how='all')
```
